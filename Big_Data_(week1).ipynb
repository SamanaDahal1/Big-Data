```python
#fibonacci
a=0
b=1
n=int(input("Enter num of terms: "))
for i in range (n):
  print(a)
  c=a+b
  a=b
  b=c

```

```python
#factorial
num=int(input("Enter a number: "))
a=1
for i in range (1, num + 1):
  a = a* i
  print(a)

```

```python
h=int(input("Enter height: "))
for i in range(1, h + 1):
  print('*'*i)
```

```python
def calculator():
  a=int(input("Enter first number: "))
  b=int(input("Enter second number: "))
  c=input("Enter operator: ")
  if c=="+":
    print(a+b)
  elif c=="-":
    print(a-b)
  elif c=="*":
    print(a*b)
  elif c=="/":
    print(a/b)

calculator()

```

```python
a = ["a", "b", "w", "e", "a", "b", "d", "f"]
b = []
for item in a:
    if item in b:
        a.remove(item)
        break
    b.append(item)
print(a)


```

```python
sample_dict = { "name": "Anderson", "age": 35, "salary": 5000, "city": "London"}
keys = ["name", "salary"]
new = {k: sample_dict[k] for k in keys if k in sample_dict}
print(new)
```
